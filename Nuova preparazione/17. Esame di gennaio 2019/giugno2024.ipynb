```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import scipy.stats as st
import statsmodels.api as sm
import math
```

```python
def density(x,a):
    return 6/a**3*((a*x)-(x**2))
```

```python
a = 12
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 12))
x = np.linspace(0, a,100)



ax1.plot(x, density(x,a))
ax2.plot(x, (((3*x**2)/(a**2))-((2*x**3)/(a**3))))
plt.show()

y = pd.Series([density(val,a) for val in x])
```

```python
#print(type(y[0]))
sm.qqplot(y, fit=True, line="45")
plt.show()
```
