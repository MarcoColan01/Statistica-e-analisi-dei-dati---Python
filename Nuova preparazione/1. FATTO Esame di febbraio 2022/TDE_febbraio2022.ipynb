```python
#4. Tracciate un grafico, diverso dal boxplot, che secondo voi ben rappresenta la distribuzione dell'attributo consumo. 
print('Consumo è un attributo quantitativo continuo, per cui la rappresentazione più appropriata è un istogramma.')
auto.consumo.hist()
plt.show()
```

```python
''' 
5. I dati evidenziano una relazione tra il consumo e la cilindrata di un'auto? In caso affermativo, di che tipo è tale relazione? 
Motivate la vostra risposta anche sullabase di un'opportuna visualizzazione grafica e sul calcolo di un indice numerico.
'''
print('Per visualizzare la relazione tra due attributi usiamo uno scatter plot')
auto.plot.scatter('consumo','cilindrata')
plt.show()
print(f'In questo caso lo scater plot evidenzia una relazione di tipo indiretta tra i due attributi. L indice di correlazione pari a {auto.consumo.corr(auto.cilindrata)} conferma tale ipotesi.')
```
