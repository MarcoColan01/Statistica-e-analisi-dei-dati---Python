```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.constants import golden
import statsmodels.api as sm
from scipy.stats import norm

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize=(5.0, 5.0/golden))

heroes = pd.read_csv('../heroes.csv', sep = ';', index_col=0)
```

```python
#INDICE DI GINI
```

```python
#Vediamo un primo grafico che traccia l'indice di Gini per due valori distinti
def gini_2_val(f):
    return 1- f**2 - (1-f)**2
x = np.arange(0,1.01, .01)
y = list(map(gini_2_val,x))
plt.plot(x,y)
plt.ylim((0,0.55))
plt.show()
print('La funzione assume punti di min e max in corrispondenza dei casi di minima e massima eterogeneità')
```

```python
#Funzione per calcolare l'indice di Gini sui valori di una serie
def gini(series):
    return 1- (sum(series.value_counts(normalize = True)
                  .map(lambda f: f**2)))
```

```python
#Applichiamo la funzione appena scritta agli attributi dei supereroi
publisher = heroes[heroes['Publisher'].isin(['Marvel Comics', 'DC Comics'])]['Publisher']
eye_color = heroes[pd.notnull(heroes['Eye color'])]['Eye color']
hair_color = heroes[pd.notnull(heroes['Hair color'])]['Hair color']

print('Publisher:', gini(publisher))
print('Eye color:', gini(eye_color))
print('Hair color:', gini(hair_color))
print('Nome eroe:', gini(heroes.index))
print('L attributo più omogeneo è quello relativo all editore')
```

```python
#Indice di Gini normalizzato
def generalized_gini(series):
    freq = series.value_counts(normalize = True)
    s = len(freq)
    return (1-(sum(freq.map(lambda f: f**2)))) * s / (s-1)

print('Publisher:', generalized_gini(publisher))
print('Eye color:', generalized_gini(eye_color))
print('Hair color:', generalized_gini(hair_color))
print('Nome eroe:', generalized_gini(heroes.index))
```

```python
#ENTROPIA
```

```python
#Andamento della funzione dell'entropia
x = np.arange(0.01, 1.01, 0.01)
y = list(map(lambda f: -1 * np.log2(f),x))
plt.plot(x,y)
plt.show()
```

```python
#Altro andamento
def entropy_2_val(f):
    return 0 if f in (0,1) else -f * np.log2(f) - (1-f) * np.log2(1-f)

x = np.arange(0, 1.01, .01)
y = list(map(entropy_2_val, x))
plt.plot(x,y)
plt.ylim((0, 1.1))
plt.show()
```

```python
#Calcolo degli indici di entropia 
def entropy(series):
    return sum((series.value_counts(normalize = True)
               .map(lambda f: -f * np.log2(f))))

print('Publisher:', entropy(publisher))
print('Eye color:', entropy(eye_color))
print('Hair color:', entropy(hair_color))
print('Nome eroe:', entropy(heroes.index))
```
