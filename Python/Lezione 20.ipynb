```python
#scipy.stats è un modulo che ci aiuta a calcolare le funzioni di massa di probabilità, di densità e di ripartizione.
#Per ognuna delle distribuzioni che abbiamo visto, esiste una classe apposita
import scipy.stats as st

Z = st.norm(0, 1)     # Z è una variabile aleatoria normale. ATTENZIONE: IL SECONDO PARAMETRO è LA DEVIAZIONE STANDARD
Z.pdf(0)              #Funzione di densità di probabilità
```

```python
#Verifichiamo la forma a campana del grafico di densità di probabilità di una normale
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,100)   #parametri: estremo sinistro, estremo destro e intervallo dei valori. Restituisce un array
#PDF capisce se è un solo numero o un array e, in base a ciò, fa i calcoli giusti.
y = Z.pdf(x)
plt.plot(y)
plt.show()
```

```python
Z.cdf(0)         #Funzione di distribuzione cumulativa (fi)
```

```python
[2* Z.cdf(n) -1 for n in (1,2,3)]    #probabilità che un'osservazione di una v.a. ricada al più in n deviazioni standard
```

```python
def approx (n):
    x = np.linspace(-10, 10, n+1)
    y = Z.pdf(x)
    return sum(y * 20/n)

approx(1000)
```
