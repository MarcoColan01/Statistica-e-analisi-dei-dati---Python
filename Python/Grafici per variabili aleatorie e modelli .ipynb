```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creare un DataFrame di esempio
data = {'variabile_discreta': [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]}
df = pd.DataFrame(data)

# Calcolare la funzione di ripartizione cumulativa (CDF)
values, counts = np.unique(df['variabile_discreta'], return_counts=True)
cdf = np.cumsum(counts) / np.sum(counts)

# Visualizzare la funzione di ripartizione
plt.step(values, cdf, where='post')
plt.title('Funzione di Ripartizione Cumulativa (CDF)')
plt.xlabel('Valori della Variabile Discreta')
plt.ylabel('Probabilità Cumulativa')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creare un DataFrame di esempio
data = {'variabile_discreta': [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]}
df = pd.DataFrame(data)

# Calcolare la massa di probabilità
values, counts = np.unique(df['variabile_discreta'], return_counts=True)
pmf = counts / np.sum(counts)

# Visualizzare la massa di probabilità con un grafico a bastoncini
plt.stem(values, pmf, basefmt='k', linefmt='b-', markerfmt='bo')
plt.title('Massa di Probabilità della Variabile Aleatoria Discreta (Grafico a Bastoncini)')
plt.xlabel('Valori della Variabile Discreta')
plt.ylabel('Probabilità')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creare un DataFrame di esempio
data = {'variabile_continua': [1.5, 2.3, 3.1, 2.8, 3.5, 4.2, 3.9, 4.7, 5.4, 6.1]}
df = pd.DataFrame(data)

# Calcolare la funzione di ripartizione cumulativa (CDF)
values = np.sort(df['variabile_continua'])
cdf = np.linspace(0, 1, len(values))

# Visualizzare la funzione di ripartizione cumulativa (CDF) con un grafico a linea
plt.plot(values, cdf, drawstyle='steps-post')
plt.title('Funzione di Ripartizione Cumulativa (CDF) della Variabile Aleatoria Continua')
plt.xlabel('Valori della Variabile Continua')
plt.ylabel('Probabilità Cumulativa')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Creare un DataFrame di esempio
data = {'variabile_binomiale': [0, 1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Definire i parametri della distribuzione binomiale
n = 5  # numero di tentativi
p = 0.5  # probabilità di successo in ogni tentativo

# Calcolare la funzione di ripartizione cumulativa (CDF)
values = np.sort(df['variabile_binomiale'])
cdf = binom.cdf(values, n, p)

# Visualizzare la funzione di ripartizione cumulativa (CDF) con un grafico a linea
plt.step(values, cdf, where='post')
plt.title('Funzione di Ripartizione Cumulativa (CDF) della Variabile Aleatoria Binomiale')
plt.xlabel('Numero di Successi')
plt.ylabel('Probabilità Cumulativa')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Creare un DataFrame di esempio
data = {'variabile_binomiale': [0, 1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Definire i parametri della distribuzione binomiale
n = 5  # numero di tentativi
p = 0.5  # probabilità di successo in ogni tentativo

# Calcolare la funzione di massa di probabilità (PMF)
values = np.arange(0, n + 1)
pmf = binom.pmf(values, n, p)

# Visualizzare la funzione di massa di probabilità (PMF) con un grafico a bastoncini
plt.stem(values, pmf, basefmt='k', linefmt='b-', markerfmt='bo')
plt.title('Funzione di Massa di Probabilità (PMF) della Variabile Aleatoria Binomiale')
plt.xlabel('Numero di Successi')
plt.ylabel('Probabilità')
plt.show()
```
