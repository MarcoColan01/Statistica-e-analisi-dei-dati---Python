```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creare un DataFrame di esempio
data = {'variabile_discreta': [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]}
df = pd.DataFrame(data)

# Calcolare la funzione di ripartizione cumulativa (CDF)
values, counts = np.unique(df['variabile_discreta'], return_counts=True)
cdf = np.cumsum(counts) / np.sum(counts)

# Visualizzare la funzione di ripartizione
plt.step(values, cdf, where='post')
plt.title('Funzione di Ripartizione Cumulativa (CDF) di una Variabile Aleatoria Discreta')
plt.xlabel('Valori della Variabile Discreta')
plt.ylabel('Probabilità Cumulativa')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creare un DataFrame di esempio
data = {'variabile_discreta': [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]}
df = pd.DataFrame(data)

# Calcolare la massa di probabilità
values, counts = np.unique(df['variabile_discreta'], return_counts=True)
pmf = counts / np.sum(counts)

# Visualizzare la massa di probabilità con un grafico a bastoncini
plt.stem(values, pmf, basefmt='k', linefmt='b-', markerfmt='bo')
plt.title('Massa di Probabilità della Variabile Aleatoria Discreta (Grafico a Bastoncini)')
plt.xlabel('Valori della Variabile Discreta')
plt.ylabel('Probabilità')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Creare un DataFrame di esempio
data = {'variabile_binomiale': [0, 1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Definire i parametri della distribuzione binomiale
n = 5  # numero di tentativi
p = 0.5  # probabilità di successo in ogni tentativo

# Calcolare la funzione di ripartizione cumulativa (CDF)
values = np.sort(df['variabile_binomiale'])
cdf = binom.cdf(values, n, p)

# Visualizzare la funzione di ripartizione cumulativa (CDF) con un grafico a linea
plt.step(values, cdf, where='post')
plt.title('Funzione di Ripartizione Cumulativa (CDF) della Variabile Aleatoria Binomiale')
plt.xlabel('Numero di Successi')
plt.ylabel('Probabilità Cumulativa')
plt.show()



n = 20
p = 0.3
x = range(20)
X = st.binom(n,p)
X.pmf(x)
plt.title('Funzione di massa di probabilità della Variabile Aleatoria Binomiale')
plt.vlines(x,0, X.pmf(x))
plt.plot(x, X.pmf(x), 'o')
plt.show()
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom
from ipywidgets import *

# Creare un DataFrame di esempio
data = {'variabile_geometrica': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Definire il parametro della distribuzione geometrica
p = 0.3  # probabilità di successo in ogni prova

# Calcolare la funzione di ripartizione cumulativa (CDF)
values = np.arange(1, max(df['variabile_geometrica']) + 1)
cdf = geom.cdf(values, p)

# Visualizzare la funzione di ripartizione cumulativa (CDF) con un grafico a gradino (plt.step)
plt.step(values, cdf, where='post')
plt.title('Funzione di Ripartizione Cumulativa (CDF) della Variabile Aleatoria Geometrica')
plt.xlabel('Numero di Tentativi')
plt.ylabel('Probabilità Cumulativa')
plt.show()

p = 0.4
x = range(20)
X = st.geom(p)
X.pmf(x)
plt.title('Funzione di massa di probabilità della Variabile Aleatoria Geometrica')
plt.vlines(x,0, X.pmf(x))
plt.plot(x, X.pmf(x), 'o')
plt.show()

```

```python
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import poisson
# Creare un DataFrame di esempio
data = {'variabile_poisson': [0, 1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Definire il parametro della distribuzione di Poisson
lambda_param = 2.5  # parametro di intensità

# Calcolare la funzione di ripartizione cumulativa (CDF)
values = np.arange(0, max(df['variabile_poisson']) + 1)
cdf = poisson.cdf(values, lambda_param)

# Visualizzare la funzione di ripartizione cumulativa (CDF) con un grafico a gradino (plt.step)
plt.step(values, cdf, where='post')
plt.title('Funzione di Ripartizione Cumulativa (CDF) della Variabile Aleatoria di Poisson')
plt.xlabel('Numero di Eventi')
plt.ylabel('Probabilità Cumulativa')
plt.show()


l = 100
x = range(100)
X = st.poisson(l)
#X.pmf(x)
plt.title('Funzione di massa di probabilità della Variabile Aleatoria di Poisson')
plt.vlines(x,0, X.pmf(x))
plt.plot(x, X.pmf(x), 'o')
plt.show()
```

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)


x = np.linspace(uniform.ppf(0.01),
                uniform.ppf(0.99), 100)
plt.title('Funzione di densità di probabilità di una Variabile Aleatoria Uniforme Continua')
ax.plot(x, uniform.pdf(x),
       'r-', lw=5, alpha=0.6, label='uniform pdf')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Creare un DataFrame di esempio
data = {'variabile_esponenziale': [1.0, 2.5, 3.8, 4.2, 5.7]}
df = pd.DataFrame(data)

# Definire il parametro della distribuzione esponenziale
lambda_param = 0.5  # parametro di tasso

# Calcolare la funzione di ripartizione cumulativa (CDF)
values = np.sort(df['variabile_esponenziale'])
cdf = 1 - np.exp(-lambda_param * values)

# Visualizzare la funzione di ripartizione cumulativa (CDF) con un grafico a linea
plt.plot(values, cdf)
plt.title('Funzione di Ripartizione Cumulativa (CDF) della Variabile Aleatoria Esponenziale')
plt.xlabel('Valori della Variabile Esponenziale')
plt.ylabel('Probabilità Cumulativa')
plt.show() 



# Calcolare la funzione di densità di probabilità (PDF)

values = np.linspace(0, max(df['variabile_esponenziale']), 100)
pdf = lambda_param * np.exp(-lambda_param * values)

# Visualizzare la funzione di densità di probabilità (PDF) con un grafico a linea
plt.plot(values, pdf)
plt.title('Funzione di Densità di Probabilità (PDF) della Variabile Aleatoria Esponenziale')
plt.xlabel('Valori della Variabile Esponenziale')
plt.ylabel('Densità di Probabilità')
plt.show()
```
