```python
#costruzione di un df a partire da un file csv
#argomento sep: come i dati sono separati nel csv
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
heroes = pd.read_csv('heroes.csv', sep = ';', index_col = 0)
heroes
```

```python
#accesso tramite nome del carattere: sottoinsieme di colonne
heroes['Gender']
```

```python
#accesso tramite slicing: sottoinsieme di righe
heroes['Agent 13':'Air-Walker']
```

```python
#accesso con loc
```

```python
#un solo valore: ritorna una serie
heroes.loc['Professor X']
```

```python
#slicing con iloc: ritorna un df
heroes.iloc[42:46] #42 incluso 46 escluso
```

```python
#loc selezionando solo alcune righe con slicing
heroes.loc['Professor X', 'Height':'Weight']
```

```python
#metodo at: accede a un elemento
heroes.at['Superman', 'Strength']
```

```python
#metodo iat: come prima ma con indici
heroes.iat[500, -1]
```

```python
'''
metodo sort_values: ordina gli elementi del df secondo 
una chiave di ordinamento, un carattere del df. 
Argomento ascending = False vuol dire ordine decrescente
'''
heroes.sort_values(by = 'Weight', ascending = False)[:5]
```

```python
#metodo sort_index: ordina in base all'indice
heroes.sort_index()[-5:]
```

```python
print('Sia sort_index che sort_values restituiscono un dataframe, in particolare\nuna copia del df di partenza')
```
