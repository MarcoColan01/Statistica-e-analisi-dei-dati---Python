```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.constants import golden 

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

cars = pd.read_csv('mtcars.txt', delimiter = '\t', decimal=",")

cars
```

```python
#6.1 Tracciare l'istogramma del tempo impiegato a fare 400 metri
cars['test400metri'].hist()
plt.show()
```

```python
'''
6.2 Per il tempo impiegato a fare 400 metri calcolate:
    2 indici di posizione centrale
    2 indici di dispersione
'''

print('Indici di posizione centrale: media ', cars.test400metri.mean(), 'mediana ', cars.test400metri.median())
print('Indici di dispersione: varianza ', cars.test400metri.var(), 'dev. standard ', cars.test400metri.std())
```

```python
'''
6.3 Tracciate un grafico opportuno per controllare se è plausibile che test400metri
segue una legge normale
'''
import statsmodels.api as sm
sm.qqplot(cars.test400metri.dropna(), fit = True, line = '45')
plt.show()
```

```python
#6.4 è plausibile che test400metri segua una legge normale ?
print('Sì, test400metri segue una legge normale perchè il carattere si distribuisce bene lungo la bisettrice.')
```
