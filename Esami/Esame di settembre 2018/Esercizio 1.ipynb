```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.constants import golden 

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

fin = pd.read_csv('finanziamenti.csv', delimiter = ';', decimal=",")

#fin
```

```python
#1.2 Calcolate la tabella della frequenze assolute di UNITA
unita_abs = pd.crosstab(index = fin.UNITA, columns = 'Frequenze assolute', colnames = [''])
unita_abs
```

```python
#1.3 Tracciate un grafico opportuno per descrivere UNITA
unita_abs.plot.bar()
plt.show()
```

```python
#1.4.1 Indicate quale sottoinsieme di osservazioni è stato usato
print('Sottoinsieme di osservazioni con totspese compreso tra 0 e 4 mln di euro')
```

```python
'''
1.5.1 Create una variabile progetti_a che contenga la parte di dataset relativa ai progetti per i quali
la quota provinciale di finanziamento è minore di quella regionale, e un'altra chiamata progetti_b che 
contenga la parte di dataset relativa ai progetti restanti
'''

progetti_a = fin[fin['FinProvincia'] < (fin['FinRegione'])]
progetti_b = fin[fin['FinProvincia'] >= (fin['FinRegione'])]
```

```python
#1.5.2 Quanti sono i progetti di tipo A e quanti quelli di tipo B ?ù
print('Progetti A sono', len(progetti_a), 'Progetti B sono', len(progetti_b))
```

```python
#1.6.1 Selezionate i progetti A con finanziamento provinciale compreso tra 200 incluso e 1000.
app = progetti_a[progetti_a['FinProvincia'] >= 200]
selezione_progetti_a = app[app['FinProvincia'] < 1000]
selezione_progetti_a
```

```python
#1.6.2 Tracciate l'istogramma del finanziamento provinciale di tali progetti imponendo classi di ampiezza di 100 euro
bins = np.arange(0,1001,100)
selezione_progetti_a['FinProvincia'].hist(bins=bins)
plt.show()
```

```python
#1.6.3 Tracciate il boxplot per la medesima quantità
selezione_progetti_a['FinProvincia'].plot.box()
plt.show()
```

```python
#1.6.4 Tra i due grafici di prima quale trovate maggiormente informativo ?
print('Il boxplot perchè mostra anche info sulla mediana.')
```

```python
#1.6.5 Media e deviazione standard dell'importo finanziato dalla provincia
print(selezione_progetti_a['FinProvincia'].mean(), selezione_progetti_a['FinProvincia'].std())
```

```python
#1.6.6 Quanti hanno ricevuto finanziamento provinciale compreso tra 500 e 700 euro ?
app2 = selezione_progetti_a[selezione_progetti_a['FinProvincia'] > 500]
len(app2[app2['FinProvincia'] < 700])
```

```python
#1.6.7 Calcolate un indice numerico e un grafico per evidenziare la relazione tra finanziamento provinciale e spese sostenute
selezione_progetti_a.plot.scatter('FinProvincia', 'TotSpese')
n = selezione_progetti_a['FinProvincia'].corr(selezione_progetti_a['TotSpese'])
print(n)
plt.show()
selezione_progetti_a
```

```python
#1.6.8 Rimuovete i progetti che si discostano notevolmente dall'andamento generale
selezione_progetti_a = selezione_progetti_a[selezione_progetti_a['TotSpese'] > 2000 ]
selezione_progetti_a.plot.scatter('FinProvincia', 'TotSpese')
n = selezione_progetti_a['FinProvincia'].corr(selezione_progetti_a['TotSpese'])
print(n)
plt.show()
```
