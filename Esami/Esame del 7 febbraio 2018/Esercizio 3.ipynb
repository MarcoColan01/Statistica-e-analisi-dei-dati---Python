```python
import numpy as np 
import pandas as pd 
import scipy.stats as st
import matplotlib.pyplot as plt 
from scipy.constants import golden 

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

pesca = pd.read_csv('pesca.csv', delimiter = ',', decimal=".")
#pesca
```

```python
#3.1 peso.pescato è discreto o continuo ?
print('peso.pescato è continuo')
```

```python
#3.2 Tracciare un grafico della distribuzione di peso.pescato 
pesca['peso.pescato'].hist()
plt.show()
```

```python
#3.3 Percentuale di giorni in cui è arrivato un pescato compreso tra 10 e 15 quintali
pesca_10 = pesca[pesca['peso.pescato'] > 10]
pesca_15 = pesca_10[pesca_10['peso.pescato'] < 15]
(len(pesca_15) / len(pesca))*100
```

```python
#3.4 Parametri della variabile normale che potrebbe descrivere peso.pescato
mu = pesca['peso.pescato'].mean()
sigma = pesca['peso.pescato'].std()
print(mu, sigma)
```

```python
#3.5 Calcolare la probabilità che arrivi che arrivi giornalmente un carico compreso tra 10 e 15 quintali.
x = st.norm(loc = mu, scale = sigma)
y = x.cdf(10)
z = x.cdf(15)
z - y
```

```python
#3.8 Consideriamo forza.del.mare e tempesta. Tracciare un grafico che mostri una relazione tra essi
pesca.plot.scatter('forza.del.mare', 'tempesta')
plt.show()
```

```python
#3.9 Descrivete la relazione tra i due caratteri tenendo conto del grafico prodotto al punto 3.8
print(pesca['forza.del.mare'].corr(pesca['tempesta']))
print('Il grafico prodotto prima evidenzia come, in condizioni di tempesta, la forza del mare sia alta e viceversa,\ncome si vede anche dall indice di correlazione.')
```
