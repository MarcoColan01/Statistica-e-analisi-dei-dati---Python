```python
'''
Consideriamo ora i valori di test400metri come realizzazione campionaria di
un campione casuale estratto dalla popolazione X = tempo necessario a fare 400 metri
'''

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.constants import golden 

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

cars = pd.read_csv('mtcars.txt', delimiter = '\t', decimal=",")
cars
```

```python
#9.1 Fornire una stima del valore atteso del tempo necessario a fare 400 metri
testAtteso = cars['test400metri'].sum()/len(cars)
testAtteso
```

```python
#9.3 Lo stimatore usato è non distorto ?
print('Lo stimatore usato (media campionaria) è non distorto perchè la media campionaria è sempre non distorta rispetto al valore atteso.')
```

```python
#9.4 Stima della deviazione standard di X
cars['test400metri'].std()
```

```python
'''
9.5 Ipotizzando che il tempo necessario per fare 400 metri segua una legge normale, calcolare la 
probabilità che, nello stimare il valore atteso di X, si commetta un errore al più uguale a 0.25, 
in eccesso o in difetto. 
'''
import scipy.stats as st
dev = cars['test400metri'].std()
n = len(cars.test400metri.dropna())
z = st.norm()
2*z.cdf(0.25*(n**0.5)/dev) -1
```
