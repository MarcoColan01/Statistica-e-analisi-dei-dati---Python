```python
import numpy as np
import matplotlib.pyplot as plt

lambda_exp = 0.1
p_geom = 0.1

x_exp = np.linspace(0, 10, 1000)
cdf_exp = 1 - np.exp(-lambda_exp * x_exp)

k_geom = np.arange(0, 11)
cdf_geom = 1 - (1 - p_geom) ** k_geom

plt.plot(x_exp, cdf_exp, label='Esponenziale(0.1)')
plt.plot(k_geom, cdf_geom, 'ro-', label='Geometrica(0.1)')
plt.xlabel('Valore')
plt.ylabel('Funzione di distribuzione cumulativa')
plt.title('CDF di variabili aleatorie esponenziale e geometrica')
plt.legend()
plt.grid(True)
plt.show()
```
