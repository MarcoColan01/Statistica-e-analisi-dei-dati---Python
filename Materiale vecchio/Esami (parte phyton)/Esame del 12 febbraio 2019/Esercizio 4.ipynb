```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.constants import golden 
import statsmodels.api as sm

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

cars = pd.read_csv('carsharing.csv', delimiter = ';', decimal=",")
carsnp = cars[cars.RushHour == 0]
#carsnp
```

```python
#4.1 Tracciate un grafico per descrivere la distanza percorsa negli orari non di punta
plt.subplot(1,2,1)
carsnp.Distance.hist()
plt.subplot(1,2,2)
carsnp.Distance.plot.box()
plt.show()
```

```python
#4.2 È plausibile affermare che negli orari non di punta la distanza segue una legge normale?
sm.qqplot(carsnp.Distance, fit = True, line = '45')
plt.show()
print('Negli orari non di punta, la distanza segue una legge normale, come si vede dalle misurazioni che si distribuiscono bene sulla bisettrice.')
```
