```python
'''
6.11 Fissato a = 0.85, determinare l'errore massimo commesso con probabilità >= ad a, per
eccesso o per difetto, nella stima del valore atteso di X.
'''
import scipy.stats as st
import math

X = st.norm()
n = len(cars)
sigma = cars.trasmissione.std()
a = ((0.85+1)/2)
phi = X.ppf(a)
e = (phi*sigma)/math.sqrt(n)
e
```
