```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.constants import golden 
import scipy.stats as st

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

cani = pd.read_csv('cani.csv', sep = ';', decimal=",", parse_dates=True)
```

```python
#3.1 EDVI e Allodiast sono indipendenti ? Motivate la risposta con anche un grafico di supporto
cani.plot.scatter('EDVI', 'Allodiast')
plt.show()
print('Come mostra il grafico, esiste una relazione diretta tra EDVI e Allodiast')
print(cani.Allodiast.corr(cani.EDVI))
print('Indice di correlazione tendente ad 1 conferma tale ipotesi.')
```

```python
#3.2 Si può dire che Allodiast segue una legge normale ?
cani['Allodiast'].hist()
plt.show()
print('Il grafico sopra conferma come Allodiast segua una legge normale')
```

```python
#3.3 Controllate che media e mediana di Allodiast siano circa uguali
print(cani.Allodiast.mean(), cani.Allodiast.median())
```
