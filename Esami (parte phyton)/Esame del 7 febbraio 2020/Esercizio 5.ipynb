```python
'''
punto 1. calcolare la media di cilindrata
punto 2. calcolare la deviazione standard di cilindrata
punto 3. 
punto 4.
punto 5.
punto 6.
punto 7.
punto 8.
'''

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.constants import golden 

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

cars = pd.read_csv('mtcars2.txt', sep = '\t', index_col = 0, decimal=",", dtype ={
    'consumo':np.float64, 'cilindrata':np.float64, 'test400metri':np.float64
})

cilindrate = cars['cilindrata']
cilindrate

'''punto 1'''
avgCyldr = 0
count = 0

for y in cilindrate:
    avgCyldr += y 
    count+=1
    
avgCyldr /= count 
avgCyldr
```

```python
'''punto 2'''
cilindrate.std()
```

```python
'''punto 3'''
from scipy.stats import norm 
valoriSimulati= norm.rvs(loc=avgCyldr, scale = cilindrate.std(), size=32)
```

```python
'''punto 4'''
levels = np.linspace(0,1,100)

plt.plot(cilindrate.quantile(levels), pd.Series(valoriSimulati).quantile(levels), 'o')
plt.show()
```

```python
'''punto 5'''
valoriSimulatiSorted = valoriSimulati.sort()
valoriSimulatiSorted 
```

```python
'''punto 6'''
cilindrateSorted = cilindrate.sort_values()
cilindrateSorted
```

```python
'''punto 7'''
plt.plot(pd.Series(valoriSimulati).quantile(levels), cilindrateSorted.quantile(levels), 'o')
plt.show()
```
