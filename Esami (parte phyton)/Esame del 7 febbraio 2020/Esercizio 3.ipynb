```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.constants import golden 

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

cars = pd.read_csv('mtcars.txt', sep = '\t', index_col = 0, decimal=",", dtype ={
    'consumo':np.float64, 'cilindrata':np.float64, 'test400metri':np.float64
})

'''punto 2'''
cilindrata = cars['cilindrata']
cilindrata.plot.box()
plt.show()

'''punto 4'''
print(cilindrata.quantile(.25))
print(cilindrata.quantile(.50))
print(cilindrata.quantile(.75))

'''punto 5'''
dist_interquantile = cilindrata.quantile(.75) - cilindrata.quantile(.25)
print('Distanza interquantile: ', dist_interquantile)

'''punto 6'''
modelli = cars['peso']
levels = np.linspace(0,1,100)
plt.plot(modelli.quantile(levels), cilindrata.quantile(levels), 'o')
plt.show()


```
