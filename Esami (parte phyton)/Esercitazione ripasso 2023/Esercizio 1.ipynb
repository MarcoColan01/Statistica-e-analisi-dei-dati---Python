```python
#Esercizio 1
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
#Punto 1.2 Fissiamo ora lambda = 5
```

```python
#1.2.1 Tracciare il grafico della massa di probabilità di X 
l = 5
x = range(15)
X = st.poisson(l)
plt.vlines(x,0,X.pmf(x))
plt.plot(x, X.pmf(x), 'o')
plt.show()
```

```python
#1.2.2 Probabilità che P assuma valori maggiori di 6
#Idea: P(X > 6) = 1-P(X <= 6)
1-X.cdf(6)
```
