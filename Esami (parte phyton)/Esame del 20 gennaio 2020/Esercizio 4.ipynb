```python
# 7.  Utilizzate le info contenute nelle tabelle prodotte nel punto precedente per rispondere alle seguenti domande:
# 7.1 Quanti sono i modelli di auto a 5 marce con trasmissione automatica (=0)?
marcetras_abs.iloc[2,0]
```

```python
# 7.2 Quanti sono i modelli di auto a 5 marce con trasmissione manuale ?
marcetras_abs.iloc[2,1]
```

```python
# 7.3 Qual'è la percentuale di modelli con 5 marce e trasmissione automatica ?
(marcetras_abs.iloc[2,0] / marcetras_abs.sum().sum())*100
```

```python
# 7.4 Qual'è la percentuale di modelli che hanno 5 marce e trasmissione manuale ?
(marcetras_abs.iloc[2,1] / marcetras_abs.sum().sum())*100
```

```python
# 7.5 Qual'è la percentuale di modelli con 4 marce ?
(marcetras_abs.iloc[1].sum() / marcetras_abs.sum().sum())*100
```
