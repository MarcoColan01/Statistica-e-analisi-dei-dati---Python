```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.constants import golden 

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

cars = pd.read_csv('mtcars.txt', delimiter = '\t', decimal=",")
```

```python
#8.1 Tracciare un grafico opportuno per controllare se esiste una relazione tra cilindrata e test400metri
cars.plot.scatter('cilindrata', 'test400metri')
plt.show()
```

```python
#8.2 Come cambia il tempo per fare 400 metri all'aumentare della cilindrata ?
print('All aumentare della cilindrata diminuisce il tempo per fare 400 metri')
```
