```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import scipy.stats as st
from scipy.constants import golden 
import statsmodels.api as sm

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize = (5.0, 5.0/golden)) 

astici = pd.read_csv('astici.csv', sep = ',', decimal=".", parse_dates=True)
#Concentriamoci ora su kg.di.pesce
```

```python
#2.1 Tracciare la funzione cumulativa empirica di kg.di.pesce
ecdf = sm.distributions.ECDF(astici['kg.di.pesce'])
x = np.arange(astici['kg.di.pesce'].min(), astici['kg.di.pesce'].max()+1)
y = ecdf(x)
plt.step(x,y)
plt.show()
```

```python
#2.2 Stimare valore atteso e varianza di kg.di.pesce
e = astici['kg.di.pesce'].mean()
vr = astici['kg.di.pesce'].var()
print('Valore atteso:', e, '\nVarianza:', vr)
```

```python
#2.3 Gli stimatori usati precedentemente sono non distorti ?
print('La media campionaria è sempre stimatore non distorto del valore atteso, mentre per la varianza va verificato.')
```
